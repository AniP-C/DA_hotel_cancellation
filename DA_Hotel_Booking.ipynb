```python
# we have found that the reservation date is in object type but we always perform analysis on numbers so we will comvert object data type to date time 
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'] , format="%d/%m/%Y")
```

```python
# this checks which coloumn is null and then returns the sum of  it.

df.isnull().sum()

#  as we can seewe have 488 coiury coloumns as NULL and 4 children coloumn as NULL 
```

```python
#  Now we will remove the company and agent's data as this is of no use and also it'll be pain in head to correct the data

df.drop(['company' , 'agent'] , axis = 1 , inplace=True)
#  now this line will remove company and agent coloumn axis = 1 means we wanna remove a coloumn , inplace = True means we wanna change the original data frame

# Now we'll remove country rows with missing values
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
#  it means we are taking adr values which is rate per night less than 5000 only as it is kinda outlier
df = df[df['adr'] < 5000]
```

#Data Analysic and Visualization

```python
# .value_counts(normalize = True) it returns the cancellation percentage
# category name and how many times it is present in the colomn , if we do normalize = True then it returns percentage
cancelled_per = df['is_canceled'].value_counts(normalize=True)
print(cancelled_per)
```

```python
#  to plot the figue
plt.figure(figsize=(5,4))
plt.title("Reservation status count")
plt.bar(['Not Cancelled' , 'Cancelled'], df['is_canceled'].value_counts())
```

```python
#  now we will see which hotel's cancellation rate is higher 
plt.figure(figsize=(8,4)) # this fun .figure(figsize) helps in sizing the graph

# we are making chart with using seabon library
ax1 = sns.countplot(x = 'hotel' , hue='is_canceled' , data = df , palette='Blues')
legend_labels = ax1.get_legend_handles_labels()

# ax1.legend(bbox_to_anchor(1,1))
plt.title("Reservation status in different hotels" , size = 20)
plt.xlabel("hotel")
plt.ylabel("number of reservations")

plt.show()
```

```python
# we are filtering the cancellation hotel wise , in city hotel and resort hotels
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
#  we are trying to find the effect of price on cancellation of the hotels
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()

city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("Avg Daily Rate in city and resort hotel" , fontsize = 30)
plt.plot(resort_hotel.index , resort_hotel['adr'] , label = "Resort Hotel")
plt.plot(city_hotel.index , city_hotel['adr'] , label = "City Hotel")
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x = 'month' , hue = 'is_canceled' , data = df  )

legend_labels = ax1.get_legend_handles_labels()

# ax1.legend(bbox_to_anchor(1,1))
plt.title("Reservation per months" , size = 20)
plt.xlabel("month")
plt.ylabel("number of reservations")
plt.legend(['not canceled' , 'canceled'])
plt.show()
```

```python
# how pricing is related to the cancellation

plt.figure(figsize=(15,8))
plt.title('ADR per month' , fontsize = 30)
# sns.barplot('month' , 'adr' , data = df[df['is_canceled'] ==1].groupby('month')[['adr']].sum().reset_index())
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

# plt.legend(fontsize = 20)
plt.show()
```

```python
# cancellation rate on the basis of countries
cancelled_data = df[df['is_canceled'] ==1 ]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("Top 10 countries with reservation canceled")

plt.pie(top_10_country, autopct='%.2f' , labels=top_10_country.index)
plt.show()
```

```python
df['market_segment'].value_counts()
```

```python
# percentage of customers comes from following segment
df['market_segment'].value_counts(normalize=True)*100
```

```python
cancelled_data['market_segment'].value_counts(normalize=True)*100
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date' , inplace=True)

# not cancelled data 
not_canceled_data = df[df['is_canceled'] == 0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date' , inplace=True)

plt.figure(figsize=(20,6))
plt.title('Avg Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'] , not_canceled_df_adr['adr'] , label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'] , cancelled_df_adr['adr'] , label = 'cancelled')
plt.legend()
```

```python
# filtering the data from 2016 to 2017 september
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016') & (cancelled_df_adr['reservation_status_date'] < '2017-09')]
not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date'] > '2016') & (not_canceled_df_adr['reservation_status_date'] < '2017-09')]

```

```python
#  plotting the plot for the new data frame
plt.figure(figsize=(20,6))
plt.title('Avg Daily Rate' , fontsize= 30)
plt.plot(not_canceled_df_adr['reservation_status_date'] , not_canceled_df_adr['adr'] , label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'] , cancelled_df_adr['adr'] , label = 'cancelled')
plt.legend()
plt.show()
```
